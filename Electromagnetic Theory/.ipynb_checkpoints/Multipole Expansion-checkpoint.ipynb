```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm  #Color maps
```

```python
def dipole_pot(X,Y,mu=1):
    theta = np.arctan2(Y,X)
    r2 = X**2 + Y**2
    Z = mu*np.cos(theta) / r2
    return Z

def log_(Z):
    M = Z*(Z>0)
    N = Z*(Z<=0)
    return np.log(M+0.01)-np.log(-N+0.01)

N = 100
a, b = -1, 1
t = np.linspace(a,b,N)
X, Y = np.meshgrid(t,t)
Z = dipole_pot(X,Y)
plt.imshow(log_(Z), cmap=cm.jet)
```
