```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('datos_equipotenciales.xlsx')
```

```python
data
```

```python
voltaje_1x=data['X.1'][0:4]
voltaje_2x=data['X.1'][4:8]
voltaje_3x=data['X.1'][8:12]
voltaje_4x=data['X.1'][12:16]
voltaje_1y=data['Y.1'][0:4]
voltaje_2y=data['Y.1'][4:8]
voltaje_3y=data['Y.1'][8:12]
voltaje_4y=data['Y.1'][12:16]
voltaje_ajuste_x=[0.05, -0.1, -0.05, 0.1]
voltaje_ajuste_y=[2, -2, 6, -6]
```

```python
    plt.figure(figsize=(16,6))
    plt.scatter(voltaje_1x,voltaje_1y,label='Voltaje de -2')
    plt.scatter(voltaje_2x,voltaje_2y,label='Voltaje de 2')
    plt.scatter(voltaje_3x,voltaje_3y,label='Voltaje de -1')
    plt.scatter(voltaje_4x,voltaje_4y,label='Voltaje de 1')
    plt.scatter(voltaje_ajuste_x,voltaje_ajuste_y,label='Voltaje de 0')
    plt.legend(fontsize = 15)
    plt.grid(True)
    plt.title('Lineas Equipotenciales Barra-Barra', fontsize = 18)
    plt.ylabel('Posición vertical', fontsize = 15)
    plt.xlabel('Posición horizontal', fontsize = 15)
    plt.xticks(size = 15)
    plt.yticks(size = 15)
    plt.xlim(-7,7)
    plt.ylim(-9,9)
    plt.show()
```
