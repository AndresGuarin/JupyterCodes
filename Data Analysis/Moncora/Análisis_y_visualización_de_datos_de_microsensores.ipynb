## **Presion**

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.style.use("fivethirtyeight") # Establecer el estilo de gráfico
```

```python
## Recopilar datos
estacion = 'mE1_00007'           # Nombre estacion miniEva
start = '2023-04-30 00:00:00'    # Fecha de inicio
end   = '2023-05-07 23:00:00'    # Fecha de fin

data = MakeSens.download_data(estacion, start, end)  # Descargar datos en un set de datos
data.index = pd.DatetimeIndex(data.index)            # Convertir a tipo datetime el índice
dat = data.resample('D').mean() # Crear conjunto de datos con reducción a datos por día por promedio

dat_std = data.resample('D').std() # Crear conjunto de datos con reducción a datos por día por desviación estandar
```

```python
dat['pm25_1_AE'].plot(ax=ax, label='Concentración Material Particulado PM2.5',
                   marker='o', markersize=8) 
```

```python
def grafica(ax,dat,dat_std,columna,label,titulo,xlabel,ylabel):
    """
    @params
        ax : axis
             Eje de una figura
        dat : dataframe
              Set de datos
        dat_std : dataframe
              Set de datos de desviación estándar
        columna : str
                  Nombre de la columna de datos
        label : str
                Etiqueta del conjunto de datos
        titulo : str
                 Nombre del título del grafico
        xlabel : str
                 Etiqueta eje x
        ylabel : str
                 Etiqueta eje y
    @ejemplo
        >> fig, ax = plt.subplots(figsize=(12,6))
        >> label = 'Concentración Material Particulado PM2.5'
        >> titulo = 'Instituto Técnico Damazo Zapata'
        >> xlabel = 'Tiempo'
        >> ylabel = 'PM2.5 $[\mu g / m^3]$'
        >> grafica(ax, dat, dat_std, 'pm25_1_AE', label, titulo,xlabel,ylabel)
        >> plt.show()
    """

    ## Graficar datos 
    dat[columna].plot(ax=ax, label=label,marker='o', ms=8) 
    plt.fill_between(dat.index, dat[columna]+dat_std[columna],dat[columna]-dat_std[columna],alpha=0.5,color='red')

    ## Agregar etiquetas a cada punto de datos
    for x, y in zip(dat.index, dat[columna]):
        ax.text(x, y+1, f"{y:.1f}", ha="center", va="bottom", 
                fontsize=12, fontweight="bold") #ha: alineamiento horizontal, va: alineamiento vertical

    ## Configurar título, etiquetas, leyenda y límite del eje 'y'
    ax.legend() # Leyenda
    ax.set_title(titulo, fontsize=24, fontweight="bold")      # Titulo
    ax.set_xlabel(xlabel, fontsize=18, fontweight="bold")     # Etiqueta eje x
    ax.set_ylabel(ylabel, fontsize=18, fontweight="bold")     # Etiqueta eje y
    ax.set_ylim(-dat[columna].max()*0.4, dat[columna].max() * 1.6)              # Límites eje y

    ## Convertir las fechas de las etiquetas del eje x a formato Mes-día
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b-%d'))
    ax.xaxis.set_major_locator(mdates.DayLocator())
    plt.xticks(rotation=45)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
label = 'Concentración Material Particulado PM2.5'
titulo = 'Instituto Técnico Damazo Zapata'
xlabel = 'Tiempo'
ylabel = 'PM2.5 $[\mu g / m^3]$'
grafica(ax, dat, dat_std,'pm25_1_AE', label, titulo,xlabel,ylabel)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
label = 'Humedad'
titulo = 'Instituto Técnico Damazo Zapata'
xlabel = 'Tiempo'
ylabel = 'Humedad $[\%RH]$'
grafica(ax, dat, dat_std,'humedad', label, titulo,xlabel,ylabel)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
label = 'Presión'
titulo = 'Instituto Técnico Damazo Zapata'
xlabel = 'Tiempo'
ylabel = 'Presión $[hPa]$'
grafica(ax, dat, dat_std, 'presion', label, titulo,xlabel,ylabel)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
label = 'Temperatura'
titulo = 'Instituto Técnico Damazo Zapata'
xlabel = 'Tiempo'
ylabel = 'Temperatura $[°C]$'
grafica(ax, dat, dat_std, 'temperatura', label, titulo,xlabel,ylabel)
plt.show()
```
