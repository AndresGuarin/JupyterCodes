## Importación de librerias importantes

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Guardando datos

```python
# Casos de colisiones elásticas
momentums1 = [[11.8, 0.6, 16.3],
             [2.0, 4.0, 3.5],
             [3.5, 3.7, 7.4],
             [3.8, 3.0, 1.9],
             [9.0, 13.1, 11.5]]

energia_cin1 = [[8.5, 40.4, 5.7],
               [12.2, 7.6, 12.9],
               [11.0,11.4,18.1],
               [11.8, 14.2, 14.1],
               [17.2, 24.5, 21.7]]


# Casos de colisiones plásticas
momentums2 = [[5.0, 7.9, 6.6],
             [7.2, 10.3, 16.5],
             [16.7, 18.6, 9.1],
             [17.8, 5.7, 22.5],
             [11.2, 11.5, 2.2]]

energia_cin2 = [[53.9, 56.5, 55.4],
               [45.6, 49.2, 55.8],
               [65.9, 67.4, 59.5],
               [74.6, 57.7, 77.1],
               [60.5, 60.7, 52.1]]

# Datos de perdida de energía
# Colisiones elásticas
PerdidaK1=[[60.1, 1.6, 75.9],
             [26.1, 40.4, 46.9],
             [43.8, 45.3, 62.8],
             [19.3, 13.7, 9.1],
             [100, 100, 100]]

# Colisiones plásticas
PerdidaK2=[[10.3,15.9,13.4],
             [23.9,33.4,52.2],
             [29.3, 32.4, 16.6],
             [19.0,7.0,23.6],
             [21.1,21.8,4.3]]
           

casos = [r'$m_1>m_2$', r'$m_1>>m_2$',r'$m_1<m_2$',r'$m_1<<m_2$',r'$m_1=m_2$',]
```

# Graficando datos

```python
plt.figure(figsize=(10,4))

ncasos = [int(i) for i in range(5)]

formato = ['o','s','^']
colors = ['red', 'blue', 'orange']
labels = ['Prueba 1', 'Prueba 2', 'Prueba 3']

plt.subplot(121)
for j in range(3):
    plt.plot(ncasos, [momentums1[i][j] for i in range(5)], formato[j], color=colors[j], label=labels[j], markersize=7.5)

#Ponemos un título y ajustamos el tamaño y pocisión de la leyenda
plt.title('Colisión elástica', fontsize = 14)
plt.legend(fontsize = 10, ncol=1)
#Agregamos un nombre y tamño de letra a las etiquetas de los ejes
plt.xlabel('Casos', fontsize = 12)
plt.ylabel('$\%Error \ p$', fontsize = 12)
#Modificamos el tamaño de los valores en los ejes
plt.xticks(ticks = ncasos ,labels = casos, size = 10)
plt.yticks(size = 14)
#plt.ylim(1.4, 27)
    
plt.subplot(122)
for j in range(3):
    plt.plot(ncasos, [momentums2[i][j] for i in range(5)], formato[j], color=colors[j], label=labels[j], markersize=7.5)

#Ponemos un título y ajustamos el tamaño y pocisión de la leyenda
plt.title('Colisión plástica', fontsize = 14)
plt.legend(fontsize = 10, ncol=1)
#Agregamos un nombre y tamño de letra a las etiquetas de los ejes
plt.xlabel('Casos', fontsize = 12)
plt.ylabel('$\%Error \ p$', fontsize = 12)
#Modificamos el tamaño de los valores en los ejes
plt.xticks(ticks = ncasos ,labels = casos, size = 10)
plt.yticks(size = 14)
plt.ylim(1.4, 27)

plt.savefig('momentums.png', format='png') #Se guarda la figura en el computador
plt.show() #mostramos todo lo que graficamos
```
